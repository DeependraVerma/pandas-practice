```python
song = soup.find_all("td", class_ = "clamp-summary-wrap")
```

```python
len(song)
```

```python
name = []
artist = []
date_of_release = []
summary = []
metascore = []
user_rating = []
for i in song:
    name.append(i.find_all("h3")[0].text.strip())
    artist.append(i.find_all("div", class_ = "artist")[0].text.strip())
    date_of_release.append(i.find_all("span")[0].text.strip())
    summary.append(i.find_all("div",class_ = "summary")[0].text.strip())
    try:
        metascore.append(i.find_all("div",class_ = "metascore_w large release positive")[0].text)
    except:
        metascore.append(np.nan)
    try:    
        user_rating.append(i.find_all("div",class_ = "metascore_w user large release positive")[0].text)
    except:
        user_rating.append(np.nan)
        
dfs = {"Name of the Song": name, "Artist": artist, "Date of Release":date_of_release,"Description":summary,"Metascore":metascore,"User Score":user_rating}

df= pd.DataFrame(dfs)
```

```python
df_final = pd.DataFrame()
for j in range(0,41):
    url = "https://www.metacritic.com/browse/albums/genre/date/rock?page={}".format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webhtml = requests.get(url,headers = headers).text
    soup = BeautifulSoup(webhtml,"lxml")
    song = soup.find_all("td", class_ = "clamp-summary-wrap")
    
    name = []
    artist = []
    date_of_release = []
    summary = []
    metascore = []
    user_rating = []
    for i in song:
        name.append(i.find_all("h3")[0].text.strip())
        artist.append(i.find_all("div", class_ = "artist")[0].text.strip())
        date_of_release.append(i.find_all("span")[0].text.strip())
        summary.append(i.find_all("div",class_ = "summary")[0].text.strip())
        try:
            metascore.append(i.find_all("div",class_ = "metascore_w large release positive")[0].text)
        except:
            metascore.append(np.nan)
        try:    
            user_rating.append(i.find_all("div",class_ = "metascore_w user large release positive")[0].text)
        except:
            user_rating.append(np.nan)
        dfs = {"Name of the Song": name, "Artist": artist, "Date of Release":date_of_release,"Description":summary,"Metascore":metascore,"User Score":user_rating}
        df= pd.DataFrame(dfs)
        df_final = df_final.append(df,ignore_index=True)
```
