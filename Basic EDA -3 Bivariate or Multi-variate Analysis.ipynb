```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
tittanic = pd.read_csv("train.csv")
```

```python
flights = sns.load_dataset("flights")
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

# 1. ScatterPlot (Numerical-Numerical column)

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], hue = tips["sex"],style=tips["smoker"],size=tips["size"])
plt.show()
```

# 2. BarPlot - Categorical- Numerical Column

```python
plt.figure(figsize=(10,10))
sns.barplot(x = tittanic["Pclass"], y = tittanic["Age"])
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = tittanic["Pclass"], y = tittanic["Fare"], hue= tittanic["Sex"])
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = tittanic["Pclass"], y = tittanic["Fare"], hue= tittanic["Sex"])
plt.show()
```

# 3. BoxPlot Categorical- Numerical Column

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = tittanic["Sex"], y = tittanic["Age"], hue = tittanic["Survived"])
plt.show()
```

# 4. DistPlot - (Numerical - Categorical Data)

```python
plt.figure(figsize=(10,10))
sns.distplot(tittanic[tittanic["Survived"]==0]["Age"], hist=False)
sns.distplot(tittanic[tittanic["Survived"]==1]["Age"],hist=False)
plt.show()
```

# 5. Heatmap (Categorical - Categorical Dataset)

```python
sns.heatmap(pd.crosstab(tittanic["Survived"], tittanic["Pclass"]),annot=True,fmt='g')
```

# 6. ClusterMap (Categorical - Categorical Dataset)

```python
sns.clustermap(pd.crosstab(tittanic["Parch"],tittanic["Survived"]))
```

# 7. Pairplot 

```python
sns.pairplot(iris, hue = "species")
```

# 8. Lineplot (x-axis = Time based(Date, month, year etc.), y- axis = dependent on time)

```python
new = flights.groupby("year").sum().reset_index()
```

```python
sns.lineplot(x = new["year"], y= new["passengers"])
```

```python
sns.heatmap(flights.pivot_table(index="month", columns="year", values = "passengers"))
```

```python
sns.clustermap(flights.pivot_table(index="month", columns="year", values = "passengers"))
```
