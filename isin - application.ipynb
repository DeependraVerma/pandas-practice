```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("deliveries.csv")
```

```python
df
```

```python
df.columns
```

```python
death_over = df[df["over"]>15]
```

```python
death_over
```

```python
all_batsman= death_over.groupby("batsman")["ball"].count()
```

```python
all_batsman
```

```python
x = all_batsman>200
```

```python
x
```

```python
batsman = all_batsman[x]
```

```python
batsman_s = batsman.index.to_list()
```

```python
batsman_s
```

```python
final = df[df["batsman"].isin(batsman_s)]
```

```python
runs = final.groupby("batsman").sum()
```

```python
total_runs = runs["batsman_runs"].sort_values(ascending = False)
```

```python
balls = final.groupby("batsman").count()
```

```python
total_balls = balls["ball"].sort_values(ascending = False)
```

```python
strike_rate = (total_runs/total_balls)*100
```
