```python
def batsman_runs(batsman):
    name = deliveries[deliveries["batsman"] == batsman]
    name = name[["bowling_team","batsman_runs"]]
    data = name.groupby("bowling_team").sum().sort_values(by = "batsman_runs",ascending = False).head(1)
    return data.index[0]
```
