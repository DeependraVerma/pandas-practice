```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
```

```python
df = deliveries[deliveries["batsman_runs"]==6]
```

```python
df
```

```python
six = df.pivot_table(index = "over", columns="batting_team", values = "batsman_runs", aggfunc="count")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,15))

sns.heatmap(six, annot = True)

plt.show()
```
