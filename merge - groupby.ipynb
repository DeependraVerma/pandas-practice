```python
import pandas as pd
import numpy as np
```

```python
deliveries = pd.read_csv('deliveries.csv')
```

```python
deliveries
```

```python
match = pd.read_csv("matches.csv")
```

```python
match
```

```python
df = deliveries.merge(match,how="inner", left_on="match_id", right_on="id")
```

```python
df
```

```python
new_df = df.groupby(["season","batsman"])["batsman_runs"].sum().reset_index()
```

```python
new_df = new_df.sort_values(by = "batsman_runs",ascending = False)
```

```python
new_df.drop_duplicates(subset="season", keep="first").sort_values(by = "season", ascending=False)
```
